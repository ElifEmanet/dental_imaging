# Mean Squared Error and Modified z-Score of Test Images

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Get the mse and modified z-score of test images and their corresponding binary classification:

```python
mse_array = np.load("../test_results/mse16-07-2022_21:47:40.npy")
bin_class = np.load("../test_results/test_bin16-07-2022_21:47:40.npy")
cl_new = np.load("../test_results/test_true_class16-07-2022_21:47:40.npy")
```

Plot histogramm for MSE (binary classes):

```python
normal = mse_array[bin_class==0]
anomaly = mse_array[bin_class==1]

fig, ax = plt.subplots(figsize=(6,6))

ax.hist(normal, bins=50, density=False, label="normal", alpha=.6, color="green")
ax.hist(anomaly, bins=50, density=False, label="anomaly", alpha=.6, color="red")

plt.title("Distribution of the Reconstruction Loss")
plt.legend()
plt.show()
```

Plot histogramm for MSE ("classification_new" classes):

```python
zero = mse_array[cl_new==0]
one = mse_array[cl_new==1]
two = mse_array[cl_new==2]
eight = mse_array[cl_new==8]
ten = mse_array[cl_new==10]
eleven = mse_array[cl_new==11]

fig, ax = plt.subplots(figsize=(6,6))

ax.hist(zero, bins=50, density=False, label="0", alpha=.6, color="green")
ax.hist(one, bins=50, density=False, label="1", alpha=.6, color="red")
ax.hist(two, bins=50, density=False, label="2", alpha=.6, color="blue")
ax.hist(eight, bins=50, density=False, label="8", alpha=.6, color="cyan")
ax.hist(ten, bins=50, density=False, label="10", alpha=.6, color="black")
ax.hist(eleven, bins=50, density=False, label="11", alpha=.6, color="orange")

plt.title("Distribution of the Reconstruction Loss")
plt.legend()
plt.show()
```
