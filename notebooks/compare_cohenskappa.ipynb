# Notebook to visualize the results for all methods with AE and VAE

```python
import matplotlib.pyplot as plt
import numpy as np
```

Results for the AE using MSE method:

```python
epoch = np.array([10, 20, 30, 40])
ck_2 = np.array([0.02094, 0.0109, 0.0123, 0.006621])
ck_5 = np.array([0.01853, 0.019999, 0.0022, 0.01806])
ck_10 = np.array([0, 0.05146, 0.03419, 0.01189])

```

```python
plt.figure(figsize=(8, 5), dpi=300)
plt.plot(epoch, ck_2, color = 'b', marker='o', label = "Latent Dimension: 2")
plt.plot(epoch, ck_5, color = 'm', marker='o', label = "Latent Dimension: 5")
plt.plot(epoch, ck_10, color = 'g', marker='o', label = "Latent Dimension: 10")

plt.xlabel("Epoch")
plt.ylabel("Cohen's Kappa Score")
plt.title("AE: Cohen's Kappa Score for Average Training Set MSE Method")
plt.legend()
plt.show()
```

Now for the VAE:

```python
ck_2_v = np.array([0.02086, 0.01035, 0.01823, 0.01268])
ck_5_v = np.array([0.04121, 0.01308, 0.03589, 0.05996])
ck_10_v = np.array([0.009884, 0.02755, 0.05831, 0.009884])
```
