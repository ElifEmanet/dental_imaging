Plot the calculated pdf values of test images

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Get the pdf values of test images and their corresponding classification (both binary and multiclass):

```python
probs = np.load("../test_results/test_prob16-07-2022_21:47:40.npy")
cl_new = np.load("../test_results/test_true_class16-07-2022_21:47:40.npy")
bin_class = np.load("../test_results/test_bin16-07-2022_21:47:40.npy")
```

```python
normal = np.where(bin_class == 0)
anomaly = np.where(bin_class == 1)
```

```python
fig, ax = plt.subplots()
ax.scatter(normal[0], probs[normal], c='y', label="normal")
ax.scatter(anomaly[0], probs[anomaly], c='b', label="anomaly")
ax.legend()
plt.show()
```
